```python
import ndjson
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('strokes.ndjson') as f:
    data = ndjson.load(f)
```

```python
df = pd.DataFrame()
for i,item in enumerate(data):
    print(f'this is {i}th loop')
    tmp = pd.DataFrame(item['events'])
    tmp.columns = ['key','eventtype','time']
    tmp['id'] = item['id']
    tmp['writer'] = item['writer']
    tmp['kana'] = item['kana']
    tmp = tmp[['id','writer','kana','key','eventtype','time']]
    df = pd.concat([df,tmp]).reset_index(drop=True)

# TAKE TOO LONG TIME
```
